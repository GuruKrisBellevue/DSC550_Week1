```python
# downloading the libraries required for the assignment
import pandas as pd
import numpy as np
import os
```

#### 1. Load the dataset as a Pandas data frame.

```python
# Getting the current working directory and the file path
path=os.getcwd()
# Creating a variable for the path where Video game file is located
videogame_file=path+'\\Video_Games_Sales_as_at_22_Dec_2016.csv'
# Reading the Videogame file into a Dataframe using read_csv method
vdo_game_df=pd.read_csv(videogame_file)
```

#### 2. Display the first ten rows of data.

```python
# Displaying the top 10 rows of the dataframe
display(vdo_game_df.head(10))
```

#### 3.Find the dimensions (number of rows and columns) in the data frame. What do these two numbers represent in the context of the data?

```python
# Using the shape function to print the number of rows and columns in the Dataframe
print(f'There are {vdo_game_df.shape[0]} Rows in the Video game sales Dataset and {vdo_game_df.shape[1]} columns')
```

##### ***What does the numbers represent?:*** Each row in the dataset represents sales figure of a Video game and its details such as Year released, Publisher, sales in millions along with the ratings. So the numbers represents there are 16719 such sales figures and 16 such attributes that describes the Video game sales.

#### 4.Find the top five games by critic score.

```python
# Using sort_values method to sort the dataframe results on the Critic_score column and arranging in
# Descending order to display top 5 rows
vdo_game_df.sort_values(by=["Critic_Score"],ascending=False).head(5)
```

#### 5.Find the number of video games in the data frame in each genre.

```python
# value_counts can be used to determine the number of video games in each genre.
vdo_game_df['Genre'].value_counts()
```

```python
# Other way of doing is by using groupby on Genre column and getting the counts and displaying the 
# results  sorted in descending order
vdo_game_df.groupby('Genre')['Name'].count().sort_values(ascending=False)
```

#### 6.Find the first five games in the data frame on the SNES platform.

```python
# Using filters based on Platform whose value is SNES and displaying first 5 rows
vdo_game_df_SNES=vdo_game_df[vdo_game_df['Platform']=="SNES"]
vdo_game_df_SNES.head(5)
```

#### 7.Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.

```python
# Grouping the data based on Publisher and getting the sum of Global sales. Then results
# are sorted in descending order to print the top 5 selling Publishers 
vdo_game_df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(5)
```

#### 8. Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.

```python
# New dataframe is created by creating copy of original dataframe
vdo_game_df1=vdo_game_df.copy()
```

```python
# New column NA Sales Percent is added by calculating the percent using the formula
vdo_game_df1['NA_Sales_Pct']=(vdo_game_df1['NA_Sales']/vdo_game_df1['NA_Sales'].sum())*100
# Displaying first five rows of the new dataframe
vdo_game_df1.head(5)
```

#### 9. Find the number NaN entries (missing data values) in each column.

```python
# Printing the number of nulls in each column using isnull function
vdo_game_df.isnull().sum()
```

```python
# Looping through each column in the dataframe and printing the number of nulls in the column
for col in vdo_game_df.columns:
    print(f"There are {vdo_game_df[col].isnull().sum()} nulls in {col} Column")
```

#### 10.Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
# New dataframe is created by creating copy of original dataframe
vdo_game_df2=vdo_game_df.copy()
# The values in the column User_score are converted to number and if theer are errors, they are 
# stored as NaN using the coerce option. 
vdo_game_df2['User_Score']=vdo_game_df2['User_Score'].apply(pd.to_numeric, errors='coerce')
```

```python
# Printing the number of Nulls in the column User Score 
print(f"Number of Nulls in the User_score column is {vdo_game_df2['User_Score'].isnull().sum()}")
```

```python
# Calculating the median afyer Non numeric values are treated
usr_score_median=vdo_game_df2['User_Score'].median()
print(f"The median of User_Score column after treating Non-Numeric values and Nulls is {usr_score_median}")
```

```python
# Using fillna method to fill the nulls with the median user score value
vdo_game_df2['User_Score']=vdo_game_df2['User_Score'].fillna(usr_score_median)
# Printing the number of Nulls in the column User Score after replacing them with Median values
print(f"Number of Nulls in the User_score column is {vdo_game_df2['User_Score'].isnull().sum()}")
```
